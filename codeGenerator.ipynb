```python
# imports

import os
import io
import sys
from dotenv import load_dotenv
from openai import OpenAI
import google.generativeai
import anthropic
from IPython.display import Markdown, display, update_display
import gradio as gr
import subprocess
```

```python
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
model='llama3.2'
```

```python
system_message = "You are an assistant that implements programming code for the given user prompt. "
system_message += "Respond only with Python code; use comments sparingly and do not provide any explanation other than occasional comments. "
system_message += "always respond with most efficient code only. do not provide multiple variations of the code."
system_message += "decline any other requests other than code generation requests"
```

```python
def chat(message, history):
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    response = ollama.chat.completions.create(model=model, messages=messages)
   
    return response.choices[0].message.content

```

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```
