```python
# imports

import os
import io
import sys
from dotenv import load_dotenv
from openai import OpenAI
import google.generativeai
import anthropic
from IPython.display import Markdown, display, update_display
import gradio as gr
import subprocess
```

```python
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
model='llama3.2'
```

```python
system_message_generator = "You are an assistant that implements programming code for the given user prompt. "
system_message_generator += "Respond only with Python code; use comments sparingly and do not provide any explanation other than occasional comments. "
system_message_generator += "always respond with most efficient code only. do not provide multiple variations of the code."
system_message_generator += "decline any other requests other than code generation requests"
```

```python
system_message_validator = "You are an assistant that validates given programming code. you accespt only programming code as user prompt"
system_message_validator += "After validating the code, you shall respond either safe to execute or not safe to execute"
system_message_validator += "decline any other requests other than code validation requests."
```

```python
def code_generator(message):
    messages = [{"role": "system", "content": system_message_generator}]  + [{"role": "user", "content": message}]
    response = ollama.chat.completions.create(model=model, messages=messages)
   
    return response.choices[0].message.content
```

```python
def code_validator(message):
    messages = [{"role": "system", "content": system_message_validator}]  + [{"role": "user", "content": message}]
    response = ollama.chat.completions.create(model=model, messages=messages)
   
    return response.choices[0].message.content
```

```python
css = """
.user_prompt {background-color: #306998;}
.codeOutput {background-color: #050;}
"""
```

```python
with gr.Blocks(css=css) as ui:
    gr.Markdown("## Generate code and validate")
    with gr.Row():
        user_prompt = gr.Textbox(label="User Prompt:", value="", lines=10)
        codeOutput = gr.Textbox(label="code output:", lines=10)
    with gr.Row():
        generate = gr.Button("Generate code")
        validate = gr.Button("Validate code")
        validationOutput = gr.Textbox(label="validation output", lines=1)
    generate.click(code_generator, inputs=[user_prompt], outputs=[codeOutput])
    validate.click(code_validator, inputs=[codeOutput], outputs=[validationOutput])

ui.launch(inbrowser=True)
```
