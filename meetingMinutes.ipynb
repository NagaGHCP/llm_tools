```python
import gradio as gr
import whisper
from transformers import pipeline
```

```python
#load wishper model
whisper_model = whisper.load_model("base")
```

```python
#load summarization pipeline model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
#process audio file and return transcirpt and summary
def process_audio(audio_file):
    # Transcribe audio
    transcription = whisper_model.transcribe(audio_file)
    transcript = transcription['text']

    # Summarize transcript
    chunks = [transcript[i:i+1000] for i in range(0, len(transcript), 1000)]
    summary = ""
    for chunk in chunks:
        result = summarizer(chunk, max_length=150, min_length=40, do_sample=False)
        summary += result[0]['summary_text'] + "\n"

    return transcript, summary
```

```python
#simple gradio interface
with gr.Blocks() as demo:
    gr.Markdown("##Meeting Minutes Generator")
    audio_input = gr.Audio(type="filepath", label="Upload English Audio")
    transcript_output = gr.Textbox(label="📄 Transcript", lines=10)
    summary_output = gr.Textbox(label="📝 Meeting Minutes", lines=10)
    submit_btn = gr.Button("Generate")

    submit_btn.click(fn=process_audio, inputs=audio_input, outputs=[transcript_output, summary_output])
```

```python
#launch Gradio
demo.launch()
```
