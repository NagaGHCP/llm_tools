```python
#Import Libraries
import os
import requests
from zipfile import ZipFile
from pydub import AudioSegment
```

```python
extract_dir = "/content/Denver_21_audio"
```

```python
#Download One Audio ZIP from MeetingBank
zip_url = "https://huggingface.co/datasets/huuuyeah/MeetingBank_Audio/resolve/main/Denver/mp3/Denver-21.zip"
zip_path = "Denver-21.zip"
```

```python
if not os.path.exists(zip_path):
    print("📥 Downloading ZIP...")
    with open(zip_path, "wb") as f:
        f.write(requests.get(zip_url).content)
```

```python
#Extract ZIP
extract_dir = "Denver_21_audio"
os.makedirs(extract_dir, exist_ok=True)

with ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"✅ Extracted to: {extract_dir}")
```

```python
#function for spliting First Audio File into 10 Clips
def split_audio_to_10_clips(file_path, segment_length_ms=5*60*1000):
    audio = AudioSegment.from_file(file_path)
    base_name = os.path.splitext(os.path.basename(file_path))[0]
    output_dir = f"{base_name}_segments"
    os.makedirs(output_dir, exist_ok=True)

    for i in range(10):
        start = i * segment_length_ms
        end = start + segment_length_ms
        if start >= len(audio):
            break
        segment = audio[start:end]
        segment.export(f"{output_dir}/{base_name}_part{i+1}.mp3", format="mp3")

    print(f"✅ Stored up to 10 clips in: {output_dir}")
```

```python
#Process First MP3 File
mp3_files = [f for f in os.listdir(extract_dir) if f.endswith(".mp3")]
if mp3_files:
    first_audio_path = os.path.join(extract_dir, mp3_files[0])
    split_audio_to_10_clips(first_audio_path)
else:
    print("⚠️ No MP3 files found in extracted ZIP.")
```
